```python
import datasets
import rich

dataset = datasets.load_dataset("nimashoghi/mptrj", split="train")
rich.print(dataset.features)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

energy_col_groups = [
    [
        "energy",
        "corrected_total_energy",
    ],
    [
        "e_per_atom_relaxed",
        "energy_per_atom",
        "ef_per_atom_relaxed",
        "ef_per_atom",
    ],
]


fig, axes = plt.subplots(1, len(energy_col_groups), figsize=(20, 5))
axes = axes.flatten()

for ax, energy_cols in zip(axes, energy_col_groups):
    for col in energy_cols:
        sns.histplot(dataset[col], ax=ax, kde=True, label=col, alpha=0.5)
    ax.set_title(f"Energy columns: {energy_cols}")
    ax.legend()
```
